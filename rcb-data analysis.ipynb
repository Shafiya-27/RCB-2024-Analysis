```python
#  RCB 2024 IPL Season Data Analysis

import pandas as pd
import matplotlib.pyplot as plt

# CSV file upload karna (Google Colab me)
from google.colab import files
uploaded = files.upload()

# Load data
df = pd.read_csv("rcb_2024_data.csv")

# Win and loss count
win_count = df[df["Result"] == "Win"].shape[0]
loss_count = df[df["Result"] == "Loss"].shape[0]

# Top batsmen runs
batsman_runs = df.groupby("Top Scorer")["Runs (Top Scorer)"].sum().sort_values(ascending=False)

# Top wicket takers
bowler_wickets = df.groupby("Top Wicket Taker")["Wickets"].sum().sort_values(ascending=False)

# Plot batsman runs
plt.figure(figsize=(8,5))
batsman_runs.plot(kind="bar", color="skyblue")
plt.title("Total Runs by Top Batsmen (RCB 2024)")
plt.ylabel("Runs")
plt.xlabel("Batsmen")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot bowler wickets
plt.figure(figsize=(8,5))
bowler_wickets.plot(kind="bar", color="lightgreen")
plt.title("Total Wickets by Bowlers (RCB 2024)")
plt.ylabel("Wickets")
plt.xlabel("Bowlers")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Win vs Loss pie chart
plt.figure(figsize=(5,5))
plt.pie([win_count, loss_count], labels=["Win", "Loss"], autopct='%1.1f%%', colors=["green", "red"])
plt.title("RCB 2024 Win vs Loss Ratio")
plt.show()
```
